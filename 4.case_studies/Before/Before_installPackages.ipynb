## Installing the packages

Do this **only once** to add the packages to the Julia distribution.

###  Method  1 
A direct and basic way using Pkg package:

```julia
using Pkg

Pkg.add("IJulia")

Pkg.add("JuMP")
Pkg.add("HiGHS")
Pkg.add("MultiObjectiveAlgorithms")

Pkg.add("Plots")
Pkg.add("LaTeXStrings")
```

###  Method  2
The fastest way using the Pkg mode:

```julia
# In the Julia REPL, press ] to enter Pkg mode (the prompt changes to pkg>). 
# Then simply type each package name after `add`:

#=

   pkg> add IJulia JuMP HiGHS MultiObjectiveAlgorithms Plots LaTeXStrings

=#  
```

### Method 3
A nice way to install something only if the package is missing:

```julia
using Pkg

packages = ["IJulia", "JuMP", "HiGHS", "MultiObjectiveAlgorithms", "Plots", "LaTeXStrings"]
installed = [p.name for p in values(Pkg.dependencies())]
to_install = filter(p -> p ∉ installed, packages)
if !isempty(to_install)
    Pkg.add(to_install)
end
```
