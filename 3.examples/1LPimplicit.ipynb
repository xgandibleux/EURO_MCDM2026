# The 1-LP model

$$
\begin{array}{llrcrrl}
\max        & z(x) = & x_1 & +     & 3x_2 &        &      \\
\text{s.t.} &        & x_1 & +     &  x_2 & \leq 14 & \qquad (1) \\
            &        &-2x_1 & +    & 3x_2 & \leq 12 & \qquad (2) \\
            &        & 2x_1 & -    &  x_2 & \leq 12 & \qquad (3) \\
            &        & x_1  & ,\, & x_2  & \geq 0  & \qquad (4)
\end{array}
$$

------

# The JuMP model

```julia
using JuMP, HiGHS
```

```julia
c = [1, 3]
d = [14, 12, 12]
T = [1 1 ; -2 3; 2 -1]
n,m = size(T)
```

```julia
modLP = Model( )

@variable(modLP, x[1:m] ≥ 0)
@objective(modLP, Max,sum(c[j]*x[j] for j=1:m))
@constraint(modLP, cst[i=1:n],sum(T[i,j]*x[j] for j=1:m) ≤ d[i])

print(modLP)
```

```julia
set_optimizer(modLP, HiGHS.Optimizer)
set_silent(modLP)
optimize!(modLP)
```

```julia
@show is_solved_and_feasible(modLP)
@show objective_value(modLP)
@show value.(x)
```
