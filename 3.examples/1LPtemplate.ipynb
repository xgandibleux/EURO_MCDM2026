# Template for setting up a JuMP model

### Create a model

Syntax : `model = Model(solver)`

```julia
using JuMP, HiGHS

modLP = Model(HiGHS.Optimizer)
```

### Define the variables

Syntax : `@variable(model, variable)`  
By default, variables are **continuous** and **unbounded**.

```julia
@variable(modLP, x1 >= 0)
@variable(modLP, x2 >= 0)
```

### Define the objective function

Syntax : `@objective(model, sense, expression)`

```julia
@objective(modLP, Max, x1 + 3x2)
```

### Define the constraints

Syntax : `@constraint(model, id, constraint)`

```julia
@constraint(modLP, cst1, x1 + x2 <= 14)
@constraint(modLP, cst2, -2x1 + 3x2 <= 12)
@constraint(modLP, cst3, 2x1 - x2 <= 12)
```

### Solve the problem

Syntaxe : `optimize!(model)`

```julia
optimize!(modLP)
```

### Extract the results

```julia
@show is_solved_and_feasible(modLP)
@show solution_summary(modLP)

@show objective_value(modLP)
@show value(x1), value(x2)
```
